```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Perceptron
from mlxtend.plotting import plot_decision_regions
```

```python
df = pd.read_csv('placement.csv')
```

```python
df.head()
```

```python
sns.scatterplot(x='cgpa',y='placement_exam_marks',data=df,hue='placed')
```

```python
x = df.drop(columns='placed',axis=1)
```

```python
x
```

```python
y = df['placed']
```

```python
y
```

```python
x_train , x_test, y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=10,shuffle=True,stratify=y)
```

```python
model = Perceptron(alpha=0.1)
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)*100 , model.score(x_test,y_test)*100
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=model)
```
